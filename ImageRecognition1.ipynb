### Pre-processing of the image 

```python
#Using CV 
```

```python
img = 'images_good//fruits.jpg'
img=cv2.imread(img)
img = np.array(img, dtype=np.uint8)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
img=cv2.resize(img,(224,224))
```

```python
resized_img=image.img_to_array(img)
final_image=np.expand_dims(resized_img,axis=0)
final_image=tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
predictions=mobile.predict(final_image)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results=imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```

```python
#Using tensorflow

mobile=tf.keras.applications.mobilenet_v2.MobileNetV2() #more accuarate 
```

```python
from tensorflow.keras.preprocessing import image 
filename = 'images_good//fruits.jpg'
img= image.load_img(filename, target_size=(224,224))
plt.imshow(img)
```

```python
resized_img=image.img_to_array(img)
final_image=np.expand_dims(resized_img,axis=0)
final_image=tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
predictions=mobile.predict(final_image)
from tensorflow.keras.applications import imagenet_utils
results=imagenet_utils.decode_predictions(predictions)
print(results)
```
